# Udemy Freebies Scraper

This is a simple example of web scraping using the selenium library for Chrome.

The aim of this notebook is to harvest URLs for course coupons from the website https://www.udemyfreebies.com/ which have may appeared over the last few days. The coupons become inactive after a period of time and therefore users would need to pay for courses if they were to enroll using an expired coupon. This script filters out such expired coupons and allows the user to select courses which they are interested in.

```python
from selenium import webdriver
import time
```

```python
#this connects to the driver and creates the variables used 

PATH = 'C:\Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(PATH)
num=1
urls=[]

#this links to the first website and then creates links to the following pages.

base = 'https://www.udemyfreebies.com/free-udemy-courses/'
websites = ['https://www.udemyfreebies.com/']


#this for loop generates the sites to visit

for r in range(2,5): #this sets the number of pages
    websites.append(base+str(r))

    
#this for loop goes through each website listed above and finds the valid coupons on them 
    
for site in range(len(websites)):
    #this connects to the website
    driver.get(websites[site])
    time.sleep(8)
    
    #this finds the button and filters out expired coupons
    
    for button in range(12):
        button_list=driver.find_elements_by_class_name("button-icon")
        try:
            if button_list[button].text == 'EXPIRED': continue
        except:
            continue
            
        #on the first run an advert appears so the driver goes back and follows the link again
        
        if num == 1:
            button_list[button].click()
            time.sleep(4)           
            driver.back()
            time.sleep(4)
            num+=1
            
        #sometimes the driver "forgets" the buttons and so these relocate them
        
        try:
            button_list[button].click()
        except:
            button_list=driver.find_elements_by_class_name("button-icon")
            button_list[button].click()
        time.sleep(4)
        
        #this grabs the URL and the coupon details and then goes back to the previous screen
        
        urls.append(driver.find_element_by_class_name("couponp-name").text)
        
        urls.append(driver.current_url)
        driver.back()
                
#this closes the driver and then prints the coupon details and URLs

time.sleep(2)
driver.quit()
[print(url) for url in urls]
```
